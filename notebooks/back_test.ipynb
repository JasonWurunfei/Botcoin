```python
from botcoin.back_test.simulator import Simulator
from botcoin.data.dataclasses.portfolio import Portfolio
```

```python
portfolio = Portfolio(cash=10000.0)
simulator = Simulator(
    portfolio=portfolio,
    required_symbols=["AAPL", "MSFT", "GOOGL"],
    benchmark_symbol="^GSPC",
    start_date="2025-01-01",
    end_date="2025-12-31",
)
simulator.load_data()
```

```python
def strategy(portfolio, date, data):
    portfolio.cash += 10  # Add $10 cash each day

simulator.strategy = strategy
```

```python
simulator.run()
```

```python
simulator.plot()
```
